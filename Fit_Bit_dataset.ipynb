**Data Transformation**


```python
daily_activity= daily_activity.rename(columns={'ActivityDate': 'Date'}) #Rename columns name Activity date to Date
hourly_steps = hourly_steps.rename(columns={'ActivityHour' : 'Time'})#Rename ActivityHour column to Time
hourly_calories = hourly_calories.rename(columns={'ActivityHour' : 'Time'})#Rename ActivityHour column to Time

```

```python
daily_activity['Date'] = pd.to_datetime(daily_activity['Date'])#change Date dattype to datetime
hourly_steps['Time'] = pd.to_datetime(hourly_steps['Time'])#change Time column to datetime

hourly_calories['Time'] = pd.to_datetime(hourly_calories['Time'])#change calories column to datetime
```

```python
daily_activity['DayOfWeek'] = daily_activity['Date'].dt.day_name()#change DayofWeek column to Date
hourly_steps['DayOfWeek'] = hourly_steps['Time'].dt.day_name()#change hourly steps to Hourly steps time
```

**MERGE HOURLY STEPS AND HOURLY CALORIES**

```python
# Select relevant columns for correlation analysis
selected_columns = ["TotalSteps", "TotalDistance", "VeryActiveDistance", "ModeratelyActiveDistance", "LightActiveDistance", "SedentaryActiveDistance", "Calories"]

# Create Corr and Plot heatmap
correlation_matrix = daily_activity[selected_columns].corr()

plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
